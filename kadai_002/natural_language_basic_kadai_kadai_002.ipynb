```python
from bs4 import BeautifulSoup
from urllib import request
from janome.tokenizer import Tokenizer
import re

#データを抽出する
url = 'https://www.aozora.gr.jp/cards/000148/files/2371_13943.html'
response = request.urlopen(url)
soup = BeautifulSoup(response,features="html.parser")
response.close()

#不要なタグを除去
main_text = main_text = soup.find('div', class_='main_text')
tags_to_delete = main_text.find_all(['rp', 'rt'])
for tag in tags_to_delete:
    tag.decompose()
#テキスト抜き出し
main_text = main_text.get_text()

# 余白、改行の除去
main_text=re.sub(r'[\u3000 \n \r]','',main_text)

# 形態素解析を用いて、品詞ごとに区切る
tokenizer = Tokenizer()
tokens = tokenizer.tokenize(main_text)

# 品詞ごとに分けてリストに格納
token_surfaces = [token.surface for token in tokens]

#ストップワード辞書の取得　及び　クリーニング
url_dictionary = 'http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/Filter/StopWord/word/Japanese.txt'
response_dictionary = request.urlopen(url_dictionary)
stopwords_list_dictionary = response_dictionary.read().decode('utf-8').split("\n")
response_dictionary.close()

#ストップワードの除去
result_text_list = [word for word in token_surfaces if word not in stopwords_list_dictionary]
print(result_text_list)
```
